# 1 способ

```python
from geopy.geocoders import Nominatim 
import folium
```

```python
addresses=['Тюмень Новатэк','Тюмень Газпром Проектирование',
           'Тюмень Лукойл','Тюмень Роснефть','Тюмень Атриум'
          ]
```

```python
tyumen_map = folium.Map(location=[57.1529, 65.5349], zoom_start=10)
geolocator = Nominatim(user_agent="Test_Tyumen")
```

```python
for address in addresses:
    # Получаем координаты адреса с помощью геокодирования библиотеки geopy
    location = geolocator.geocode(address)
    
    if location is not None:
        # Наносим маркер в виде круга
        marker = folium.Circle(
                 location=[location.latitude, location.longitude],
                 radius=25,
                 color='blue',
                 fill=True,
                 fill_color='blue',
                 fill_opacity=1  # Интенсивность
        ).add_to(tyumen_map)
        
# Отображаем карту
tyumen_map
```

# 2 Способ

Данный способ подходит, для поиска компаний по адресу через Яндекс Карты


```python
import folium
from geopy.geocoders import Nominatim
import requests
```

```python
map_moscow = folium.Map(location=[55.751244, 37.618423], zoom_start=10,tiles = "CartoDB dark_matter") # Карта Москвы
```

```python
geolocator = Nominatim(user_agent="Test_Moscow")
```

```python
# Адреса
addresses = ['140053, Московская обл, г. Котельники, мкр Силикат, дом 34',
             '117218, Москва г, Нахимовский пр-кт, дом 40',
             '125504, город Москва, вн.тер. г. Муниципальный Округ Западное Дегунино, ш Дмитровское, дом 85, этаж 2, помещение 213',
             '141100, Московская обл, г. Щёлково, пл. Ленина, дом 5, ЭТАЖ 5 КОМНАТА 501/1']
```

```python
# Функция для работы с Яндекс картами
def get_coordinates(address):
    apikey = '35dca34f-03d9-481b-ba24-e5d87ec1c731'  # Мой ключ API https://developer.tech.yandex.ru/services/
    url = f'https://geocode-maps.yandex.ru/1.x/?apikey={apikey}&format=json&geocode={address}'

    response = requests.get(url)
    response_json = response.json()

    
    # Получаем координаты из ответа
    try:
        coordinates_str = response_json['response']['GeoObjectCollection']['featureMember'][0]['GeoObject']['Point']['pos']
        longitude=float(coordinates_str[:coordinates_str.index(' ')])
        latitude=float(coordinates_str[coordinates_str.index(' ')+1:])
        return latitude, longitude
    except (KeyError, IndexError):
        return None
```

```python
for address in addresses:
    
    coordinates=get_coordinates(address)
    
    if coordinates is not None:
        # Наносим маркер
        folium.Marker(location=[coordinates[0], coordinates[1]], icon=folium.Icon(color = 'red')).add_to(map_moscow)

# Отображаем карту
map_moscow
```
